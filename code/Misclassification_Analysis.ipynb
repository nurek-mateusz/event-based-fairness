```python
def check_for_missing_values(df):
    # Initialize a dictionary to hold the count of missing values for each column
    missing_values_counts = {}
    
    # Check each column in ground_truth_labels for missing values
    for column in ground_truth_labels:
        if column in df.columns:
            # Count the number of missing (NaN) values in the column
            missing_count = df[column].isna().sum()
            missing_values_counts[column] = missing_count
        else:
            missing_values_counts[column] = "Column not in DataFrame"
    
    """
    # Display the counts of missing values
    for column, missing_count in missing_values_counts.items():
        print(f"{column}: {missing_count}")
    """
    return missing_values_counts
```

```python
def generate_dict_matrices_for_each_student(df_gt, student_id_label):
    # Initialize the dictionary
    egoid_matrices = {}
    
    # Process each egoid
    for egoid, group in df_gt.groupby(student_id_label):
        # Initialize the first matrix with NaNs to accommodate missing values
        matrix_1 = np.full((6, 6), np.nan)
        # Populate the matrix
        for i, category in enumerate(sorted(set(label.split('_')[0] for label in ground_truth_labels))):
            for j in range(1, no_surveys + 1):
                column_name = f"{category}_{j}"
                if column_name in group:
                    # Assuming one row per survey per egoid, take the first occurrence if duplicates
                    value = group[column_name].values[0] 
                    matrix_1[i, j-1] = int(value)
        
        
        # Add to the dictionary
        egoid_matrices[egoid] = np.array(matrix_1)
    return egoid_matrices

def generate_predicted_options_dataframe():
    # Dictionary to store pivoted DataFrames
    pivoted_dfs = {}

    for opinion_base in opinion_basis_list:
        # Construct the file path
        file_path = f"{base_path}\\{opinion_base}-{lambda_value}.csv"
        
        # Load the DataFrame
        df = pd.read_csv(file_path, delimiter=";", usecols=predicted_opinions_columns)
    
        # Pivot the DataFrame
        df_pivoted = df.pivot(index='StudentID', columns='SurveyNr', values='OpinionSim')
        
        # Rename the columns to follow the pattern
        df_pivoted.columns = [f"{opinion_base}_{col}" for col in df_pivoted.columns]
        
        # Reset index
        df_pivoted.reset_index(inplace=True)
        # Store in the dictionary
        pivoted_dfs[opinion_base] = df_pivoted
        
    # Initialize a variable to hold the merged DataFrame. Start with the first DataFrame in the list.
    merged_df = next(iter(pivoted_dfs.values()))
    
    # Iterate over the remaining DataFrames in the dictionary and merge them with the initial DataFrame.
    for category, pivoted_df in pivoted_dfs.items():
        if merged_df is not pivoted_df:  # Skip the first DataFrame since it's already in merged_df
            merged_df = pd.merge(merged_df, pivoted_df, on="StudentID", how="outer")
    return merged_df
```

```python
missing_values = check_for_missing_values(ground_truth)
ground_truth = ground_truth.dropna()
ground_truth.rename(columns={'egoid' :'StudentID'}, inplace=True)

predicted_opinions = generate_predicted_options_dataframe()
predicted_opinions = predicted_opinions.dropna()

common_student_ids = set(ground_truth['StudentID']).intersection(set(predicted_opinions['StudentID']))


ground_truth_aligned = ground_truth[ground_truth['StudentID'].isin(common_student_ids)]
predicted_opinions_aligned = predicted_opinions[predicted_opinions['StudentID'].isin(common_student_ids)]


print(ground_truth_aligned.head())
print(predicted_opinions_aligned.head())

gt_students_matrices = generate_dict_matrices_for_each_student(ground_truth_aligned, 'StudentID')
pred_students_matrices = generate_dict_matrices_for_each_student(predicted_opinions_aligned, 'StudentID')

def get_zipped_matrices_pred_real(gt_students, pred_students):
    zipped_matrices = {}
    for (studentID, TrueValue), (studentID2, PredValue) in zip(gt_students.items(),pred_students.items()):
        if studentID == studentID2:
            zipped_matrices[studentID] = (TrueValue, PredValue)
            print(TrueValue)
            print(PredValue)
    return zipped_matrices
zipped_matrices = get_zipped_matrices_pred_real(gt_students_matrices, pred_students_matrices)    

import numpy as np


#classify as correct if for a student more than 0.5 was classified as correct
def divide_into_classified_correctly_misclassified(zipped_matrices):
    classified_correctly = []
    misclassified = []

    for student_id, matrices in zipped_matrices.items():
        matrix1, matrix2 = matrices

        # Ensure matrices have the same shape before comparison
        if matrix1.shape == matrix2.shape:
            # Calculate the percentage of elements that are the same in both matrices
            similarity_percentage = np.mean(matrix1 == matrix2) * 100  # Element-wise comparison and mean to get the percentage

            if similarity_percentage > 50:
                classified_correctly.append(student_id)
            else:
                misclassified.append(student_id)
        else:
            # Handle the case where matrices do not have the same shape, if necessary
            # For this example, consider them misclassified
            misclassified.append(student_id)

    return classified_correctly, misclassified

classified_correctly, misclassified = divide_into_classified_correctly_misclassified(zipped_matrices)

print(len(classified_correctly))
print(len(misclassified))
#print(len(gt_students_matrices) == len(pred_students_matrices) and list(gt_students_matrices.keys())==(list(pred_students_matrices.keys())))
```
