```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import sklearn
from sklearn import metrics
```

```python
topics = ["euthanasia", "fssocsec", "fswelfare", "jobguar", "marijuana", "toomucheqrights"]
```

# naming game results 
## accuracy

```python
file_list = os.listdir("results_naming_game/")
csv_files = [file for file in file_list if file.endswith('.csv')]
```

```python
ng = pd.DataFrame({"topic":[], "average":[]})
```

```python
for topic in topics:
    for i in range(10):
        res = pd.read_csv("results_naming_game/naming_game_" + topic + str(i) + '.csv')
        accuracy = res.is_sim_survey.mean()
        ng.loc[len(ng)] = [topic, accuracy]
```

```python
plt.figure(figsize=(10, 5))
plt.grid(axis='y', linestyle='--', alpha=0.7)
sns.boxplot(x='topic', y='average', data=ng)
plt.ylabel('Average Accuracy') 
plt.title('Boxplot naming game', fontsize=16)
```

```python
averages = ng.groupby('topic')['average'].mean()
std_devs = ng.groupby('topic')['average'].std()

plt.figure(figsize=(10, 5))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.errorbar(topics, averages, yerr=std_devs, fmt='o', capsize=5, elinewidth=2, markeredgewidth=2)

plt.title('Error plot naming game', fontsize=16)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

for i, topic in enumerate(topics):
    df = pd.DataFrame({"OpinionSurvey": [], "OpinionSimulation": []})
    for j in range(10):
        res = pd.read_csv("results_naming_game/naming_game_" + topic + str(j) + '.csv')
        df = pd.concat([df, res[["OpinionSimulation", "OpinionSurvey"]]])

    confusion_matrix = metrics.confusion_matrix(np.array(df.OpinionSurvey), np.array(df.OpinionSimulation))
    normalized_confusion_matrix = confusion_matrix.astype('float') / confusion_matrix.sum(axis=1)[:, np.newaxis]

    row = i // 3
    col = i % 3
    ax = axes[row, col]
    cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=normalized_confusion_matrix)
    cm_display.plot(ax=ax)
    ax.set_title(f'topic: {topic}')
    
fig.suptitle('Naming game confusion matrices', fontsize=16)
```

## F1 score

```python
ng = pd.DataFrame({"topic":[], "average":[]})
```

```python
for topic in topics:
    for i in range(10):
        res = pd.read_csv("results_naming_game/naming_game_" + topic + str(i) + '.csv')
        accuracy = sklearn.metrics.f1_score(res.OpinionSurvey, res.OpinionSimulation, average='macro')
        ng.loc[len(ng)] = [topic, accuracy]
```

```python
plt.figure(figsize=(10, 5))
plt.grid(axis='y', linestyle='--', alpha=0.7)
sns.boxplot(x='topic', y='average', data=ng)
plt.ylabel('Average Accuracy') 
plt.title('Boxplot naming game - F1 score', fontsize=16)
```

# cogsnet results 

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))  

for i, topic in enumerate(topics):
    row = i // 3
    col = i % 3
    evaluation = pd.DataFrame({"delta":[], "average":[]})
    for j in np.arange(0.1, 1, 0.1):
        j = str(round(j,1))
        for k in range(1,11):
            res = pd.read_csv("../results/" + topic + "-" + j + "-iter" + str(k) + ".csv", sep=';')
            res["is_sim_survey"] = np.where(res['OpinionSim'] == res['OpinionSurvey'], 1, 0)
            accuracy = res.is_sim_survey.mean()
            evaluation.loc[len(evaluation)] = [j, accuracy]
    
    sns.boxplot(x='delta', y='average', data=evaluation, color="blue", ax=axes[row, col])

    axes[row, col].set_title(f'Topic: {topic}')
    axes[row, col].set_xlabel('Delta')
    axes[row, col].set_ylabel('is_sim_survey')
    axes[row, col].grid(True)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))  

for i, topic in enumerate(topics):
    row = i // 3
    col = i % 3
    evaluation = pd.DataFrame({"delta":[], "average":[]})
    for j in np.arange(0.1, 1, 0.1):
        j = str(round(j,1))
        for k in range(1,11):
            res = pd.read_csv("../results/" + topic + "-" + j + "-iter" + str(k) + ".csv", sep=';')
            res["is_sim_survey"] = np.where(res['OpinionSim'] == res['OpinionSurvey'], 1, 0)
            accuracy = res.is_sim_survey.mean()
            evaluation.loc[len(evaluation)] = [j, accuracy]
    
    averages = evaluation.groupby('delta')['average'].mean()
    std_devs = evaluation.groupby('delta')['average'].std()
    axes[row, col].errorbar(evaluation['delta'].unique(), averages, yerr=std_devs, fmt='o-', capsize=5, elinewidth=2, markeredgewidth=2, label='is_sim_survey')

    axes[row, col].set_title(f'Topic: {topic}')
    axes[row, col].set_xlabel('Delta')
    axes[row, col].set_ylabel('is_sim_survey')
    axes[row, col].grid(True)
```

```python
best_delta = pd.DataFrame({"topic":["euthanasia", "fssocsec", "fswelfare", "jobguar", "marijuana", "toomucheqrights"],
                           "delta":[0.1,0.6,0.6,0.2,0.1,0.7]})
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))  
for i, topic in enumerate(topics):
    row = i // 3
    col = i % 3
    df = pd.DataFrame({"OpinionSurvey": [], "OpinionSim": []})
    for k in range(1,11):
        delta = best_delta.delta.loc[best_delta.topic == topic].values[0]
        res = pd.read_csv("../results/" + topic + "-" + str(delta) + "-iter" + str(k) + ".csv", sep=';')
        df = pd.concat([df, res[["OpinionSim", "OpinionSurvey"]]])

    confusion_matrix = metrics.confusion_matrix(np.array(df.OpinionSurvey), np.array(df.OpinionSim))
    normalized_confusion_matrix = confusion_matrix.astype('float') / confusion_matrix.sum(axis=1)[:, np.newaxis]

    row = i // 3
    col = i % 3
    ax = axes[row, col]
    cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=normalized_confusion_matrix)
    cm_display.plot(ax=ax)
    ax.set_title(f'topic: {topic}, delta: {delta}')
fig.suptitle('CogsNet confusion matrices', fontsize=16)
```

## F1 score

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))  

for i, topic in enumerate(topics):
    row = i // 3
    col = i % 3
    evaluation = pd.DataFrame({"delta":[], "average":[]})
    for j in np.arange(0.1, 1, 0.1):
        j = str(round(j,1))
        for k in range(1,11):
            res = pd.read_csv("../results/" + topic + "-" + j + "-iter" + str(k) + ".csv", sep=';')
            accuracy = sklearn.metrics.f1_score(res.OpinionSurvey, res.OpinionSim, average='macro')
            evaluation.loc[len(evaluation)] = [j, accuracy]
    
    sns.boxplot(x='delta', y='average', data=evaluation, color="blue", ax=axes[row, col])

    axes[row, col].set_title(f'Topic: {topic}')
    axes[row, col].set_xlabel('Delta')
    axes[row, col].set_ylabel('is_sim_survey')
    axes[row, col].grid(True)
    
fig.suptitle('CogsNet F1 score', fontsize=16)
```
